```python
linguagens_nota = df['linguagens_nota'].to_numpy();
redação_nota = df['redação_nota'].to_numpy();
matemática_nota = df['matemática_nota'].to_numpy();
humanas_nota = df['humanas_nota'].to_numpy();
ciências_nota = df['ciências_nota'].to_numpy(); 

notas = np.array([linguagens_nota, redação_nota, matemática_nota, humanas_nota, ciências_nota])
```

```python
df['nota_média_final'] = 0
```

```python
df
```

```python
for index, row in df.iterrows():
  
  if row['criterio regional'] == 'Sim':
      médiaff  = row['nota_média'] + (np.dot(10 / 100, row['nota_média']))
      df.loc[index, 'nota_média_final'] = médiaff
  elif row['criterio regional'] == 'Não':
    df.loc[index, 'nota_média_final'] = row['nota_média']
```
